```python
from IPython.core.display import HTML
import pandas as pd
import numpy as np

flights_raw = pd.read_csv('D:/iamcr viz data/heatmap2.csv')
flights_raw["region"] = pd.Categorical(flights_raw["region"], flights_raw.region.unique())
flights_raw.head()
```

```python
flight_matrix = flights_raw.pivot("region", "rank group", "count")
flight_matrix
```

```python
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl

import matplotlib.pyplot as plt
%matplotlib inline


fig = plt.figure()
fig, ax = plt.subplots(1,1, figsize=(12,12))
heatplot = ax.imshow(flight_matrix, cmap='Spectral')
ax.set_xticklabels(flight_matrix.columns)
ax.set_yticklabels(flight_matrix.index)

tick_spacing = 1
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.set_title("Heatmap of Publications by Region by Journal Ranking Group")
ax.set_xlabel('Ranking Group')
ax.set_ylabel('Region')
```

```python
import seaborn as sns
fig = plt.figure(figsize=(12,12))
r = sns.heatmap(flight_matrix, cmap='coolwarm')
r.set_title("Heatmap of Publications by Region by Journal Ranking Group")
```
